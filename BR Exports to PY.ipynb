```python
# import os
# !pip install pandas 
import pandas as pd
```

```python
# saves as df data pulled from UN Comtrade Database
df = pd.read_csv('brz_exports_to_pry.csv', encoding='unicode_escape')
```

```python
# displays dataframe
df.head()
```

```python
#dropping unnecessary columns
df = df.drop('TypeCode', axis=1)
df = df.drop('FreqCode', axis=1)
df = df.drop('RefPeriodId', axis=1)
df = df.drop('RefMonth', axis=1)
df = df.drop('Period', axis=1)
df = df.drop('ReporterCode', axis=1)
df = df.drop('ReporterISO', axis=1)
df = df.drop('FlowCode', axis=1)
df = df.drop('PartnerCode', axis=1)
df = df.drop('PartnerISO', axis=1)
df = df.drop('Partner2Code', axis=1)
df = df.drop('Partner2ISO', axis=1)
df = df.drop('Partner2Desc', axis=1)
df = df.drop('ClassificationCode', axis=1)
df = df.drop('ClassificationSearchCode', axis=1)
df = df.drop('IsOriginalClassification', axis=1)
df = df.drop('IsLeaf', axis=1)
df = df.drop('CustomsCode', axis=1)
df = df.drop('CustomsDesc', axis=1)
df = df.drop('IsNetWgtEstimated', axis=1)
df = df.drop('GrossWgt', axis=1)
df = df.drop('IsGrossWgtEstimated', axis=1)
df = df.drop('MosCode', axis=1)
df = df.drop('MotCode', axis=1)
df = df.drop('MotDesc', axis=1)
df = df.drop('AltQty', axis=1)
df = df.drop('IsAltQtyEstimated', axis=1)
df = df.drop('PrimaryValue', axis=1)
df = df.drop('QtyUnitCode', axis=1)
df = df.drop('AltQtyUnitCode', axis=1)
df = df.drop('IsQtyEstimated', axis=1)
df = df.drop('AltQtyUnitAbbr', axis=1)
df = df.drop('LegacyEstimationFlag', axis=1)
df = df.drop('IsReported', axis=1)
df = df.drop('IsAggregate', axis=1)
df = df.drop('Unnamed: 47', axis=1)
df=df.drop('cif_value',axis=1)
```

```python
# renaming columns in order to provide clarity and to make merging datasets easier later
df = df.rename(columns={'RefYear':'year', 'ReporterDesc':'reporter','FlowDesc':'trade_flow','PartnerDesc':'partner','CmdCode':'hs_code','CmdDesc':'commodity', 'QtyUnitAbbr':'unit_type','Qty':'quantity','NetWgt':'net_weight','Cifvalue':'cif_value','Fobvalue':'fob_value'})
```

```python
#keeping only rows that have HS code with six digits 
df = df[df['AggrLevel']==6]
```

```python
#dropping column once all rows are with HS 6 codes 
df = df.drop(columns = 'AggrLevel')
```

```python
#displaying dataframe to see updates
df.head()
```

```python
#changing wording to make it easier to understand the direction of the trade flow 
df['trade_flow']=df['trade_flow'].replace('Export','exported_to')
```

```python
#changing unit type wording to make it easier to read / understand
df['unit_type']=df['unit_type'].replace(['1000u','thousands_of_units'],['2u','pairs_of_units'])
df['unit_type']=df['unit_type'].replace(['l','liters'],['m²','square_meters'])
df['unit_type']=df['unit_type'].replace(['m³','cubic_meters'],['u','units'])
```

```python
#checking for nulls
#interesting to see that both exports don't have any cif values at all 
df.isnull().sum()
```

```python
#saving data as new set
df.to_csv('cleaned br exports to py.csv',index=False)
```
