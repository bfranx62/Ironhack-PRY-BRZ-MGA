```python
import os
```

```python
import pandas as pd
```

```python
df = pd.read_csv('pry_exports_to_brz.csv', encoding='unicode_escape')
```

```python
df.head()
```

```python
df = df.drop(['TypeCode','FreqCode','RefPeriodId','RefMonth','Period','ReporterCode','ReporterISO','FlowCode'], axis=1)
```

```python
df.head()
```

```python
df = df.drop(['PartnerCode','PartnerISO','Partner2Code','Partner2ISO','Partner2Desc','ClassificationCode','IsNetWgtEstimated','GrossWgt','IsGrossWgtEstimated'], axis=1)
```

```python
df.head()
```

```python
df = df.drop(['ClassificationSearchCode','IsOriginalClassification','IsLeaf','AltQty','IsAltQtyEstimated','PrimaryValue','LegacyEstimationFlag','IsReported','IsAggregate'],axis=1)
```

```python
df.head()
```

```python
df=df.drop(['MosCode','QtyUnitCode','IsQtyEstimated','AltQtyUnitCode','AltQtyUnitAbbr'],axis=1)
```

```python
df.head()
```

```python
df=df.drop(['CustomsCode','CustomsDesc','MotCode','MotDesc'],axis=1)
```

```python
df.head()
```

```python
df=df.drop('Unnamed: 47',axis=1)
```

```python
df = df.rename(columns={'RefYear':'year', 'ReporterDesc':'reporter','FlowDesc':'trade_flow','PartnerDesc':'partner','CmdCode':'hs_code','CmdDesc':'commodity', 'QtyUnitAbbr':'unit_type','Qty':'quantity','NetWgt':'net_weight','Cifvalue':'cif_value','Fobvalue':'fob_value'})
```

```python
df = df[df['AggrLevel']==6]
```

```python
df = df.drop(columns = 'AggrLevel')
```

```python
df.shape
```

```python
df['trade_flow']=df['trade_flow'].replace('Export','exported_to')
```

```python
df['unit_type']=df['unit_type'].replace(['1000u','thousands_of_units'],['2u','pairs_of_units'])
```

```python
df['unit_type']=df['unit_type'].replace(['l','liters'],['m²','square_meters'])
```

```python
df['unit_type']=df['unit_type'].replace(['m³','cubic_meters'],['u','units'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop('cif_value',axis=1)
```

```python
df.to_csv('cleaned py exports to br.csv',index=False)
```
