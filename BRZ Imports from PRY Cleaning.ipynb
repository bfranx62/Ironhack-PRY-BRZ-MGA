```python
import os
import pandas as pd
```

```python
os.getcwd()
```

```python
# Saves as df data pulled from UN Comtrade Database
df = pd.read_csv("BRZ Imports from PRY.csv", encoding='unicode_escape')
```

```python
# Displays dataframe
df
```

```python
# Defines useful columns to keep
columns_to_keep = ['RefYear','ReporterDesc','FlowDesc','PartnerDesc','CmdCode','CmdDesc','AggrLevel','QtyUnitAbbr','Qty','NetWgt','Cifvalue','Fobvalue']
```

```python
# Saves the dataframe with only those columns that have useful data
df = df.loc[:,columns_to_keep]
```

```python
# Saves the dataframe keeping only those rows with an HS level of 6 digits
df = df[df['AggrLevel'] == 6]
```

```python
# Removes the now useless AggrLevel column
df = df.drop(columns= 'AggrLevel')
```

```python
# Renames columns in order to provide clarity and to make merging datasets easier later
df.rename(columns = {'RefYear':'year',
          'ReporterDesc': 'reporter',
          'FlowDesc':'trade_flow',
          'PartnerDesc':'partner',
          'CmdCode':'hs_code',
          'CmdDesc':'commodity',
          'QtyUnitAbbr':'unit_type',
          'Qty':'quantity',
          'NetWgt':'net_weight',
          'Cifvalue':'cif_value',
          'Fobvalue':'fob_value'},
          inplace=True)

# Displays new dataframe with new column names
df
```

```python
# Checks values of unit types to be updated
df['unit_type'].value_counts()
```

```python
# Creates a dictionary with old unit_type values and the values we want to change them to. 
# Updated values provide clarity in regards to unit types
replace_dict = {
    '1000u':'thousands_of_units',
    '2u':'pairs_of_units',
    'l':'liters',
    'm²':'square_meters',
    'm³':'cubic_meters',
    'u':'units'
}
```

```python
# Actually updates the unit types. 
df['unit_type'] = df['unit_type'].replace(replace_dict)
```

```python
# Displays modified values in unit_types
df['unit_type'].value_counts()
```

```python
# Changes value in trade_flow column from Import to imported_from to provide clarity of trade flows.
df['trade_flow']=df['trade_flow'].replace('Import','imported_from')
```

```python
# Saves cleaned dataframe as a new .csv file for future use. 
df.to_csv('cleaned BRZ Imports from PRY.csv',index=False)
```
